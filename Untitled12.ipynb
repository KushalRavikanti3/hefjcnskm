```python
# Import the SVC class from sklearn.svm
from sklearn.svm import SVC
# Import the confusion matrix function
from sklearn.metrics import confusion_matrix

# Split the data into training and testing sets
X = df.drop("Grad_Rate", axis=1)
y = df["Grad_Rate"]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Train the SVM model
svm = SVC(random_state=1) # Now SVC is defined and can be used
svm.fit(x_train, y_train)

# Evaluate the model
svm_score = svm.score(x_test, y_test)
print("SVM accuracy:", svm_score)

# Generate confusion matrix
y_predict_svm = svm.predict(x_test)
y_true_svm = y_test
cm_svm = confusion_matrix(y_true_svm, y_predict_svm) # Now confusion_matrix is defined

# Plot the confusion matrix
f, ax = plt.subplots(figsize=(5, 5))
sns.heatmap(cm_svm, annot=True, linewidths=0.5, linecolor="red", fmt=".0f", ax=ax)
plt.xlabel("Predicted")
plt.ylabel("True")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Assuming 'X' and 'y' are already defined as in your previous code
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(X_train, y_train) # Now X_train and y_train should have the same number of samples
```

```python
 rfc_train = rfc.predict(X_train)
 from sklearn import metrics
 print("Accuracy_Score =", format(metrics.accuracy_score(y_train, rfc_train)))

 from sklearn import metrics
 predictions = rfc.predict(X_test)
 print("Accuracy_Score =", format(metrics.accuracy_score(y_test, predictions)))

 from sklearn.metrics import classification_report, confusion_matrix
 print(confusion_matrix(y_test, predictions))
 print(classification_report(y_test,predictions))
```

```python
def encode_grad_rate(df):
  """
  This function encodes the Grad_Rate column in a DataFrame.

  Args:
      df: A Pandas DataFrame containing the Grad_Rate column.

  Returns:
      A Pandas DataFrame with the Grad_Rate column encoded as 1 if marks are below 50 and 0 otherwise.
  """

  df['Grad_Rate'] = df['Grad_Rate'].apply(lambda x: 1 if x < 50 else 0)
  return df

# Encode the Grad_Rate column in the df DataFrame
df = encode_grad_rate(df)

# Print the encoded DataFrame
print(df)
```

```python
plots = {1: [111], 2: [121, 122], 3: [131, 132, 133], 4: [221, 222, 223, 224]}
def boxplot(x, y, df):
  rows = 4
  columns = 4
  plt.figure(figsize=(7* columns, 7* rows))
  for i, j in enumerate(y):
    plt.subplot(rows, columns, i + 1)
    sns.boxplot(x=x, y=j, data=df[[x, j]], palette="Blues", linewidth=1)
    ax.set_title(j)
plt.show()
```
